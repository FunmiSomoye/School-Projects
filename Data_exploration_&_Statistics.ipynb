Stratified sampling formula

Sample size of the strata = size of entire sample / population size * layer size

                              20,000/423006 * 316417=14960.402=74.8% #calculated size of Young Adult sample
                              
                              20,000/423006 * 104761=4953.1734=24.76% #calculated size of Adult sample
                              
                              20,000/423006 * 1576=74.514383=0.3725% #calculated size of Youth sample
                              
                              20,000/423006 * 252=11.914736=0.0595% #calculated size of Senior sample

```python
#create stratified samples using figures from the formula above
ya=data2[data2['age_groupings']=='Young Adult'].sample(n=14960)#round n to nearest whole number of calculated values
a=data2[data2['age_groupings']=='Adult'].sample(n=4953)#round n to nearest whole number of calculated values
y=data2[data2['age_groupings']=='Youth'].sample(n=75)#round n to nearest whole number of calculated values
s=data2[data2['age_groupings']=='Senior'].sample(n=12)#round n to nearest whole number of calculated values
```

```python
#get stratified dataframe by using .append method on the newly created variables
first=ya.append(a)
second=first.append(y)
stratified=second.append(s)
```

In  this question we are going to implement cluster sampling. Data is divided into clusters, if it isn't already divided into clusters, and then we use simple random sampling to select a number of clusters from the sample. From the group of selected clusters, we again use simple random sampling to select random data points. These now form our sample.  

Let's implement this procedure using some randomly generated data that we are going to group into clusters.

```python
n_clusters = 10
data = scs.norm.rvs(loc=0,        # mean 0
                      scale=1,      # variance 1
                      size=100000,  # number of points
                      random_state=None)

data = pd.DataFrame({'var1': data})  # turn list into dataframe with heading var1

cluster = [rnd.randint(0,9) for c in range(100000)] # generate cluster numbers for grouping

n_col = pd.Series(cluster) # turn list into Series object
data['cluster'] = n_col.values
data.head()
```
